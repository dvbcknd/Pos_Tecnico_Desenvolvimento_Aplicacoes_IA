```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentPerformanceFactors.csv")
```

```python
df.info()
```

# **Frequencia X Nota final**

```python
covariancia = df["Attendance"].cov(df["Exam_Score"])
print(f"A corvariância de Frequencia X Nota é: {covariancia}")
```

A **Covariância** entre Frequencia e Exame é de **26.10** (arredondando). É uma covariância positiva, o que indica que há uma relação positiva entre as duas variáveis.

Isso significa que no geral, os **alunos que tem maior frequencia tendem a ter uma nota final maior.**

Abaixo temos o gráfico de dispersão:


```python
sns.scatterplot(
    data=df,
    x="Attendance",
    y="Exam_Score"
)
plt.title("Frequencia X Nota final")
plt.xlabel("Frequencia")
plt.ylabel("Nota final")
plt.show()
```

**A direção do gráfico**, que representa a relação entre Frequência e Nota final, é **positiva**. Isso significa que, **à medida que a frequência aumenta, a nota final também tende a aumentar.**




# **Horas de sono X Nota final**

```python
covariancia = df["Sleep_Hours"].cov(df["Exam_Score"])
print(f"A corvariância de Horas de sono X Nota final é: {covariancia}")
```

A **covariância** entre Horas de sono e nota final é de aproximadamente **-0.097**. Uma covariância negativa indica que há uma relação inversa entre as duas variáveis, indicando uma **correlação fraca ou quase inexistente.**

Abaixo temos o gráfico de dispersão:


```python
sns.scatterplot(
    data=df,
    x="Sleep_Hours",
    y="Exam_Score"
)
plt.title("Horas de Sono X Nota final")
plt.xlabel("Horas de Sono")
plt.ylabel("Nota final")
plt.show()
```

**A direção do gráfico**, que representa a relação entre Horas de Sono e Nota final, é **negativa**, mas muito fraca, quase zero. Embora a covariância seja ligeiramente negativa, **-0.097**, o **gráfico de dispersão mostra que as horas de sono não têm um impacto claro e consistente na nota final**, sugerindo uma correlação muito fraca ou quase inexistente.



# **Sessões de Tutoria X Nota final**

```python
covariancia = df["Tutoring_Sessions"].cov(df["Exam_Score"])
print(f"A corvariância de Sessões de Tutoria X Nota final é: {covariancia}")
```

A **covariância** entre Sessões de Tutoria e Nota final é de aproximadamente **0.75**. Uma covariância **positiva** indica que há uma relação positiva entre as duas variáveis. Isso significa que, em geral, **alunos que participam de mais essões de Tutoria tendem a ter Nota final mais altas.**

```python
sns.scatterplot(
    data=df,
    x="Tutoring_Sessions",
    y="Exam_Score"
)
plt.title("Sessões de Tutoria X Nota final")
plt.xlabel("Sessões de Tutoria")
plt.ylabel("Nota final")
plt.show()
```

**A direção do gráfico**, que representa a relação entre Sessões de Tutoria e Nota final, é **positiva**. Isso significa que, **à medida que o número de sessões de tutoria aumenta, a nota final também tende a aumentar.**



# **Heatmap de Correlação**

```python
matrix_correlacao = df[
    ["Hours_Studied",
     "Attendance",
     "Sleep_Hours",
     "Tutoring_Sessions",
     "Exam_Score"
    ]
].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    matrix_correlacao,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=.5
    )
plt.title('Heatmap de Correlação das Variáveis Numéricas')
plt.show()
```

Outras variáveis com correlação positiva, porém mais fracas, são:

Observe que Horas de Sono (Sleep_Hours) apresenta uma correlação negativa, embora seja quase zero, de -0.02.
